```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = 'data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 再畫一張histogram看分布長什麼樣子
sns.distplot(df['1stFlrSF'])
plt.show()
'''
    看起來圖形是接近normal的
    所以現在可以套用三個sigma的方式來去除離群值
'''
mu = df['1stFlrSF'].mean()
sigma = df['1stFlrSF'].std()

# 轉換前先把整數轉成浮點數才不會出錯
df[num_features] = df[num_features].astype(float)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""

# 將值限縮到3個標準差以內
df['1stFlrSF'] = df['1stFlrSF'].clip(mu - 3 * sigma, mu + 3 * sigma)

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""
keep_idx = (df['1stFlrSF'] > (mu - 3 * sigma)) & (df['1stFlrSF'] < (mu + 3 * sigma))
df, train_Y = df[keep_idx], train_Y[keep_idx]

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

## 原因解釋：

將離群值直接捨去掉之後，模型比較不會因為修剪的值而失去解釋性，

而且要fit的點比原本更少，線性模型也更容易找到systematic part in data。
