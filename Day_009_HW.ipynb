# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype(np.float64), np.dtype(np.int64)]

numeric_columns = app_train.columns[app_train.dtypes.isin(dtype_select)]

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 )])
numeric_columns.remove('SK_ID_CURR')

print("Numbers of remain columns: %d" % len(numeric_columns))

# 檢視這些欄位的數值範圍

plt.rcParams['figure.figsize'] = 20, 300

import math
print(math.ceil(len(numeric_columns)/3))

for i, col in zip(range(len(numeric_columns)), numeric_columns):
    """
    Your CODE HERE, make the box plot
    """
    plt.subplot(math.ceil(len(numeric_columns)/3), 3, i + 1)
    app_train.boxplot(column=col)

plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
import statsmodels.api as sm

plt.rcParams['figure.figsize'] = 20, 14
cdf = sm.distributions.ECDF(app_train['AMT_INCOME_TOTAL'])
plt.subplot(221)
plt.plot(cdf.x, cdf.y)
plt.xlabel('Value')
plt.ylabel('ECDF')


# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.subplot(222)
plt.plot(np.log(cdf.x, where=cdf.x!=float("-inf"), out=cdf.x), cdf.y)
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

# 自己實作不call lib的方法
"""
    解釋：
    value_counts(): unique value分別出現多少次，默認不計算NA values
    sort_index(): 將每一列按照label順序排好，由於前面value_counts返回的series其idx為unique value，等於是我們將unique value從小排到大(默認)
    cumsum(): 將series的value作累積總合，這邊的value指得是counts的部分，每一個unique value「出現的次數」的累積總合
"""
cdf = app_train.AMT_INCOME_TOTAL.value_counts().sort_index().cumsum()
plt.subplot(223)
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.subplot(224)
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
